```python
metodos_pagamento.rename(columns={
    'YearMonth': 'AnoMes',
    'quantityPix': 'Qtd_PIX',
    'valuePix': 'Valor_PIX',
    'quantityTED': 'Qtd_TED',
    'valueTED': 'Valor_TED',
    'quantityTEC': 'Qtd_TEC',
    'valueTEC': 'Valor_TEC',
    'quantityBankCheck': 'Qtd_Cheque',
    'valueBankCheck': 'Valor_Cheque',
    'quantityBrazilianBoletoPayment': 'Qtd_Boleto',
    'valueBrazilianBoletoPayment': 'Valor_Boleto',
    'quantityDOC': 'Qtd_DOC',
    'valueDOC': 'Valor_DOC'
},inplace=True)
```

```python
# Selecionando as colunas desejadas
df_selected = metodos_pagamento[['AnoMes', 'Qtd_PIX', 'Valor_PIX']]

# Exibindo as primeiras 10 linhas
print(df_selected.head(10))
```

```python
# Filtrar as linhas onde Valor_PIX > 1.000.000
df_filtrado = metodos_pagamento[metodos_pagamento['Valor_PIX'] > 1000000]

# Exibir as primeiras linhas do resultado
print(df_filtrado)
```

```python
# Criar a nova coluna
metodos_pagamento['Media_Valor_PIX'] = metodos_pagamento['Valor_PIX'] / metodos_pagamento['Qtd_PIX']

# Exibir as primeiras linhas para verificar a alteração
print(metodos_pagamento.head())
```

```python
# Converter AnoMes para string e extrair o ano
metodos_pagamento['Ano'] = metodos_pagamento['AnoMes'].astype(str).str[:4]

# Agrupar por Ano e somar os valores de PIX
df_agrupado = metodos_pagamento.groupby('Ano')['Valor_PIX'].sum().reset_index()

print(df_agrupado)

```

```python
# Ordenar o dataset pela coluna Valor_PIX em ordem decrescente
df_ordenado = metodos_pagamento.sort_values(by='Valor_PIX', ascending=False)

# Selecionar as 10 maiores transações
df_top10 = df_ordenado.head(10)
print(df_top10)
```

```python
# Filtrar os meses com mais de 1.000.000 de transações PIX
meses_acima_1M = metodos_pagamento[metodos_pagamento['Qtd_PIX'] > 1000000]

# Contar quantos meses atendem à condição
quantidade_meses = meses_acima_1M.shape[0]

# Exibir o resultado
print(quantidade_meses)
```

```python
# Calcular a média do Valor_TED
media_valor_ted = metodos_pagamento['Valor_TED'].mean()

# Calcular a mediana do Valor_TED
mediana_valor_ted = metodos_pagamento['Valor_TED'].median()

# Exibir os resultados
print(f"Média do Valor_TED: {media_valor_ted}")
print(f"Mediana do Valor_TED: {mediana_valor_ted}")
```

```python
# Converter AnoMes para string e extrair o ano
metodos_pagamento['Ano'] = metodos_pagamento['AnoMes'].astype(str).str[:4]

# Agrupar por Ano e calcular a média do Valor_Boleto
df_media_boleto = metodos_pagamento.groupby('Ano')['Valor_Boleto'].mean().reset_index()

print(df_media_boleto)
```

```python
import matplotlib.pyplot as plt

# Converter a coluna AnoMes para string caso não esteja
metodos_pagamento['AnoMes'] = metodos_pagamento['AnoMes'].astype(str)

# Ordenar os dados pelo tempo
metodos_pagamento = metodos_pagamento.sort_values(by='AnoMes')

# Criar o gráfico de linha
plt.figure(figsize=(12, 6))
plt.plot(metodos_pagamento['AnoMes'], metodos_pagamento['Valor_PIX'], marker='o', linestyle='-')

# Personalizar o gráfico
plt.xlabel("AnoMes")
plt.ylabel("Valor Total PIX")
plt.title("Evolução do Valor Total das Transações PIX ao Longo do Tempo")
plt.xticks(rotation=45)
plt.grid(True)

# Exibir o gráfico
plt.show()
```

```python
import numpy as np

# Substituir valores 0.0 por NaN na coluna Qtd_DOC
metodos_pagamento['Qtd_DOC'] = metodos_pagamento['Qtd_DOC'].replace(0.0, np.nan)

# Exibir as primeiras linhas para verificar a alteração
print(metodos_pagamento.head())
```

```python
# Calcular a correlação entre Valor_PIX e Valor_TED
correlacao = metodos_pagamento[['Valor_PIX', 'Valor_TED']].corr()

# Exibir o resultado
print(correlacao)
```

```python
# Ordenar os dados pelo tempo
metodos_pagamento = metodos_pagamento.sort_values(by='AnoMes')

# Calcular o crescimento percentual do Valor_PIX
metodos_pagamento['Crescimento_Valor_PIX'] = metodos_pagamento['Valor_PIX'].pct_change() * 100

# Calcular o crescimento percentual do Valor_TED
metodos_pagamento['Crescimento_Valor_TED'] = metodos_pagamento['Valor_TED'].pct_change() * 100

# Exibir os resultados:
# Exibir as primeiras linhas do dataframe atualizado
print(metodos_pagamento[['AnoMes', 'Crescimento_Valor_PIX', 'Crescimento_Valor_TED']].head())

```

```python
# Identificar o último mês disponível no dataset
ultimo_mes = metodos_pagamento['AnoMes'].max()

# Filtrar os dados para o último mês
dados_ultimo_mes = metodos_pagamento[metodos_pagamento['AnoMes'] == ultimo_mes]

# Selecionar apenas as colunas relevantes para os métodos de pagamento
colunas_pagamento = ['Valor_PIX', 'Valor_TED', 'Valor_TEC', 'Valor_Boleto', 'Valor_DOC']

# Calcular o valor total das transações para o último mês
total_transacoes = dados_ultimo_mes[colunas_pagamento].sum().sum()

# Calcular a participação percentual de cada método de pagamento
participacao_percentual = (dados_ultimo_mes[colunas_pagamento].sum() / total_transacoes) * 100

# Criar um dataframe com os resultados
df_participacao = participacao_percentual.reset_index()
df_participacao.columns = ['Metodo_Pagamento', 'Participacao_Percentual']

# Exibir os resultados
df_participacao
print(df_participacao)
```

```python
#salvando os resultados para análise posterior:

df_participacao.to_csv("participacao_metodos_pagamento.csv", index=False)
print("Arquivo salvo como 'participacao_metodos_pagamento.csv'")
```
